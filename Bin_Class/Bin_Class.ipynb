## **Classification Problem**

### The prdeicted value is categorical  
  
### Predict the output in True/False or Yes/No or 1/0  

### The predicted value or output is always from one of the available categories

## Example:  

- Email is Spam or Not
- Which Customer will buy life insurance
- Which Party a person is going to vote for

### The Technique that is used to solve the Classification problems is called **Logistic Regression**

## **Types of Classification**  

- **Binary Classification:**

The output is always one of the two available categories.  

**e.g.** Will customer buy life insurance (Yes/No)?  

- **Multiclass Classification:**

The output variables are more than two.  

**e.g.** Which party a person is going to vote (Democratic/Republican/Independant)?

### This curved line is called **Sigmoid / Logit function**  

## **Formula**  

The formula for this sigmoid is:  

$$sigmoid(z) = \frac{1}{1 + e^{-z}}$$  

Where:  

> $e$ : Euler's Number ~ 2.71828

We're dividing the _**1**_ by a number which is slightly greater than 1. This way the outcome will always be less than 1 and it will come up with a range between 0 to 1. If we give a set of numbers to this function, it will convert them into a range between 0 to 1. The line we'll get after using this function will be _**S**_ shaped.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[['age']], df.bought_insurance, test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
y_test
```

```python
model.score(X_test, y_test)
```

```python
model.predict_proba(X_test) # Shows probability
```
