# Principle Component Ananlysis

When solving the Machine Learning problem in real life, we ususally have hunderds of columns or features. Some of them are significantly impacting the target variables and some are much less or no impact on target. We need to identify the important features. If we get rid of these features, we'll get following two benefits:  

- Faster training and Inference  
- Data Visualization becomes easier

  
#### **PCA** is a process of figuring out most important features or principal componenets that has the most impact on the target variable.  


Few things to keep in mind before applying PCA:  

- Scale features before applying PCA:
     - For example if features in one dimension are in hundreds while in other dimension they are few, then we have to scale them first.

- Accuracy Might Drop:
     - If every features is contributing to predict target then the accurcay might get effected. It's a trade off.

```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
digits.keys()
```

```python
import pandas as pd
df = pd.DataFrame(digits.data, columns = digits.feature_names)
df.head()
```

```python
X = df
y = digits.target
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.3, random_state = 1)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
from sklearn.decomposition import PCA

pca = PCA(0.95)  #tells pca to mretain 95% of useful features

X_pca = pca.fit_transform(X)

X_pca.shape
```

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size = 0.3, random_state = 30)
```

```python
model = LogisticRegression()
model.fit(X_train_pca, y_train)
model.score(X_test_pca, y_test)
```
