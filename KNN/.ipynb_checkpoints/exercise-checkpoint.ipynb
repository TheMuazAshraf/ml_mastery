## From Sklearn.datasets load digits dataset and do the following  

#### 1. Classify digits (0 to 9) using KNN classifier. You can use different values for k neighbors and need to figure out a value of k that gives you a maximum score. You can manually try different values of k or use gridsearchcv.  

#### 2. Plot Confusion Matrix  

#### 3. Plot Classification report

```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
digits.feature_names
```

```python
digits.data
```

```python
digits.target
```

```python
digits.target_names
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(digits.data, digits.target)
df.head()
```

```python
df['target'] = digits.target
```

```python
df.head()
```

```python
X = df.drop('target', axis='columns')
y = df.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =0.3, random_state = 1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 3)
```

```python
knn.fit(X_train, y_train)
```

```python
knn.score(X_test, y_test)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import seaborn as sns
```

```python
from sklearn.metrics import confusion_matrix
y_pred = knn.predict(X_test)

cm = confusion_matrix(y_test, y_pred)
cm
```

```python
plt.figure(figsize=(7,5))
sns.heatmap(cm, annot = True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```
