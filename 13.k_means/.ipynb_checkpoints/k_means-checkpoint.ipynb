# K Means Clustering

Machine Learning algorithms are categorized into 3 types:  
1. Supervised:  
   In the given dataset, we have our class label and target variable present.  
3. Unsupervised:  
   Al we have is set of features. We don't know about target variable or class label. Using this kind of datasets, we try to identify the underlying structure in that data or sometimes try to find the clusters in that data so we can make usefull predictions out of it.  
   **K Means** is very popular clustering algorithm.

## _Working_  

The first step is to define two random points which we consider as the centers of those two clusters. We call them **_Centroids_** as well. If the k is three then we'll put three random points. The centroids can be put anywhere in the graph as shown below:  

![Random Centroids](resources/cetroids.png)  


Next step is to identify the distance of each of the datapoints from centroids. The data point near to first centroids belongs to first cluster, and datapoints newr to second centroids belong to second cluster.  

The mathematical way to do this is we draw a line connecting these centroids and then draw another line in the middle of first line which is perpendicular to each other (as shown below):  

![lines showing clusters](resources/linesinitial.png)  


Then we calculate the centre points of those datapoints in each cluster and put centroids there. Now iterate the same process until the centroids doesn't move and no more datapoints are added to any cluster. Each iteration will make clustering better.

## How to Determine the correct number of clusters (K)?  

In real cases we have so many features that makes it hard to visualize data on a scatter plot. There is a technique called **Elbow Method**.  

Before getting into Elbow method, let's first take a look at our dataset. Looking at the datapoints we take two clusters but someone might say there're three clusters and someone might say there're four and this goes on. Now it's our job to find out best possible value of **K**. For that we use Elbow method.  

The way this Elbow method works is that we start with some **K**. let's say k is equal to 2. Then we try to compute sum of square error.  

> SSE = sum of squared errors
> $$ SSE_1 = \sum_{i=0}^{n} \mathrm{dist}(x_i - c_1)^2 $$

For each of the clusters we try to compute the distance of each of the datapoint from the centroids then square it then sum it up. The above equation is for cluster one. similarly for cluster two:  

> $$ SSE_2 = \sum_{i=0}^{n} \mathrm{dist}(x_i - c_2)^2 $$

We do that for all clusters and then we get total sum of squared errors:  

> $$ SSE = SSE_1 + SSE_2 + SSE_3 + ..... + SSE_k $$

We do square just to handle negative values. Once we have all the errors, we draw a plot like following:  

![elbow plot](resources/elbow-plot.png)  


As we increase the number of clusters, the error is decreased. The best paractice is to find out an elbow in the plot. In the chart the K=4 is an Elbow point as shown below:  


![elbow point](resources/elbow-point.png)  


Whatever the datapoint this chart is representing, the good **K** value will be 4.

```python
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
%matplotlib inline
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("resources/income.csv")
df.head()
```

```python
plt.scatter(df['Age'],df['Income($)'])
```

```python
km = KMeans(n_clusters=3)
km
```

```python
y_predicted = km.fit_predict(df[['Age','Income($)']])
y_predicted
```

```python
df['cluster'] = y_predicted
df.head()
```

```python
df1 = df[df.cluster == 0 ]
df2 = df[df.cluster == 1 ]
df3 = df[df.cluster == 2 ]

plt.scatter(df1.Age, df1['Income($)'], color='green')
plt.scatter(df2.Age, df2['Income($)'], color='red')
plt.scatter(df3.Age, df3['Income($)'], color='blue')

plt.xlabel("Age")
plt.ylabel("Income($)")
plt.legend()
```

The datapoints aren't clustered correctly because there's huge difference between income scaling and age scaling. The Income scaling goes from 40,000 to 160,000 whereas Age scaling goes only 27 to 42. To eleminate this error we have to scale our features proper and for that purpose we'll use **MinMaxScaler**.

```python
scaler = MinMaxScaler()
scaler.fit(df[['Income($)']])
df['Income($)'] = scaler.transform(df[['Income($)']])
df
```

The MinMaxScaler will try to scale Income column 0 to 1.

Now Let's do the same for Age column as well.

```python
scaler.fit(df[['Age']])
df['Age'] = scaler.transform(df[['Age']])
df
```

```python
y_predicted = km.fit_predict(df[['Age','Income($)']])
y_predicted
```

```python
df['cluster'] = y_predicted
df
```

```python
km.cluster_centers_    #Centroids locations
```

```python
df1 = df[df.cluster == 0 ]
df2 = df[df.cluster == 1 ]
df3 = df[df.cluster == 2 ]

plt.scatter(df1.Age, df1['Income($)'], color='green')
plt.scatter(df2.Age, df2['Income($)'], color='red')
plt.scatter(df3.Age, df3['Income($)'], color='blue')
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], color = 'purple', marker = '+', label = 'centroids')  #showing centroids on graph
plt.xlabel("Age")
plt.ylabel("Income($)")
plt.legend()
```

```python
km.cluster_centers_    #Centroids locations
```

> Now let's try it with Elbow method

```python
k_rng = range(1,10)
sse = []
for k in k_rng:
    km = KMeans(n_clusters = k)
    km.fit(df[['Age', 'Income($)']])
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel('K')
plt.ylabel('SSE')
plt.plot(k_rng, sse)
```
