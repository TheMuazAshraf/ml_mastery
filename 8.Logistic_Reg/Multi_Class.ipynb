## **DataSet:**  

#### Scikit-Learn digits dataset

```python
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.datasets import load_digits
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
digits = load_digits()
```

```python
dir(digits)
```

```python
digits.data[25]
```

```python
plt.gray()
plt.matshow(digits.images[25])
```

```python
digits.feature_names[25]
```

```python
digits.target[25]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size = 0.2)
```

```python
X_train
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test,y_test)
```

```python
plt.matshow(digits.images[1379])
```

```python
digits.target[1379]
```

```python
model.predict([digits.data[1379]])
```

## **Confusion Matrix**  

This is used to see where the model didn't predict the correct value

```python
predicted_val = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, predicted_val)
cm
```

```python
# To Visualize this confusion matrix
import seaborn as sns
plt.figure(figsize=(10,8))
sns.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Actual')
```
